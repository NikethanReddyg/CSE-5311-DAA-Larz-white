```python
import pandas as pd
import matplotlib.pyplot as plt
import psutil
import platform
```

```python
df = pd.read_excel('SortingPerformanceResults.xlsx', sheet_name='Sorting Performance Results')
```

```python
df.head()
```

```python
# Get system information
my_system = platform.uname()
system_info = (
    f"System: {my_system.system}\n"
    f"Node Name: {my_system.node}\n"
    f"Release: {my_system.release}\n"
    f"Machine: {my_system.machine}\n"
    f"Processor: {my_system.processor}\n"
    f"CPU: {psutil.cpu_count()} cores"
)

# Get RAM information
ram_info = f"RAM: {psutil.virtual_memory().total / (1024 ** 3):.2f} GB"

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(df['arraySize'], df['bubbleSort'], label='Bubble Sort')
plt.plot(df['arraySize'], df['selectionSort'], label='Selection Sort')
plt.plot(df['arraySize'], df['insertionSort'], label='Insertion Sort')

# Add system information as text annotation
plt.text(0.02, 0.5, system_info + '\n' + ram_info, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))

plt.title('Sorting Performance')
plt.xlabel('Array Size')
plt.ylabel('Time Taken')
plt.legend()
plt.grid(True)

plt.show()
```
