```python
import pandas as pd
import matplotlib.pyplot as plt
import psutil
import platform
```

```python
df = pd.read_excel('TimeComplexityAnalysis.xlsx', sheet_name='Time Complexity Analysis')
df.head()
```

```python

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(df['n'], df['Time Taken for function1']*1000, label='Function1')
plt.plot(df['n'], df['Time Taken for function2']*1000, label='Function2')


# Add system information as text annotation
plt.text(0.02, 0.5," ", transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(facecolor='white', alpha=0.5))

plt.title('Sorting Performance')
plt.xlabel('Array Size')
plt.ylabel('Time Taken')
plt.legend()
plt.grid(True)

plt.show()
```
